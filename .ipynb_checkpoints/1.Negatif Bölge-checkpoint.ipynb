# Negative Area

Step 1 — Importing libraries.

```python
import cv2
import matplotlib.pyplot as plt
```

Step 2 — Let’s read the image.

```python
imgpath = 'Lenna.jpg'
img = cv2.imread(imgpath)
```

Step 3 — Convert the image to RGB

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

Step 4 — Let's read the image in grayscale also.

```python
gray = cv2.imread(imgpath, 0)
```

Step 5 — Let's generate the negative image.

## Algorithm of finding the negative:
* Get the red green blue values of each pixel.
* Subtract each color value from 255 and save them as new color values.

```python
colored_negative = abs(255-img)
gray_negative = abs(255-gray)
```

Step 6 — Let’s plot the results.

```python
imgs = [img, gray, colored_negative, gray_negative]
title = ['coloured', 'gray', 'coloured-negative', 'gray-negative']


plt.subplot(2, 2, 1)
plt.title(title[0])
plt.imshow(imgs[0])
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 2)
plt.title(title[1])
plt.imshow(imgs[1], cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 3)
plt.title(title[2])
plt.imshow(imgs[2])
plt.xticks([])
plt.yticks([])

plt.subplot(2, 2, 4)
plt.title(title[3])
plt.imshow(imgs[3], cmap='gray')
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
def negative_area(a,b,c,d,imgpath):
    img = cv2.imread(imgpath)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img[a:b,c:d] = abs(255-img[a:b,c:d])
    plt.imshow(img)
```

```python
negative_area(200,300,300,400,'Lenna.jpg')
```
